```python
# Importing Necessaary Libraries
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_path = "/content/download.jpeg"
img = cv2.imread(img_path)
print(img.shape)
```

```python
x = 50  # starting x coordinate of the ROI
y = 50  # starting y coordinate of the ROI
width = 100  # width of the ROI
height = 100  # height of the ROI
# Crop the image using the ROI coordinates
cropped_img = img[y:y+height, x:x+width]
# Display the cropped image using matplotlib
plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
